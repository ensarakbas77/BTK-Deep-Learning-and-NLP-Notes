```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ages = [10,20,30,30,30,40,50,60,70,75]
weights = [20,60,80,85,86,87,70,90,95,90]
```

```python
ages = np.array(ages)
weights = np.array(weights)
```

```python
plt.figure(figsize=(4,3))
plt.plot(ages, weights, "g")
plt.xlabel("yaş")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişimi")
plt.show()
```

```python
np_arr1 = np.linspace(0,10,20)
np_arr1
```

```python
np_arr2 = np_arr1 ** 2
```

```python
plt.figure(figsize=(4,3))
plt.plot(np_arr1, np_arr2, "b--") # b*-
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x) = x^2")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(np_arr1, np_arr2, "r*-")

plt.subplot(1,2,2)
plt.plot(np_arr2, np_arr1, "r")

plt.show()
```

```python
my_fig = plt.figure()
axes = my_fig.add_axes([0.1,0.1,0.3,0.3])
axes.plot(np_arr1, np_arr2, "g")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Başlık")
plt.show()
```

```python
figure = plt.figure()

ax1 = figure.add_axes([0.1, 0.1, 0.7, 0.7])
ax2 = figure.add_axes([0.3, 0.3, 0.3, 0.3])

ax1.plot(np_arr1, np_arr2, "g")
ax1.set_xlabel("x-ekseni")
ax1.set_ylabel("y-ekseni")
ax1.set_title("Ana Grafik Başlık")

ax2.plot(np_arr2, np_arr1, "b")
ax2.set_xlabel("x-ekseni")
ax2.set_ylabel("y-ekseni")
ax2.set_title("Küçük Grafik Başlık")

plt.show()
```

```python
figure = plt.figure()

# bu sayıları değiştirelim
# [x_baslangic, y_baslangic, genislik, yukseklik]
ax1 = figure.add_axes([0.1, 0.1, 0.9, 0.9])
ax2 = figure.add_axes([0.2, 0.4, 0.3, 0.3])

ax1.plot(np_arr1, np_arr2, "g")
ax1.set_xlabel("x-ekseni")
ax1.set_ylabel("y-ekseni")
ax1.set_title("Ana Grafik Başlık")

ax2.plot(np_arr2, np_arr1, "b")
ax2.set_xlabel("x-ekseni")
ax2.set_ylabel("y-ekseni")
ax2.set_title("Küçük Grafik Başlık")

plt.show()
```

```python
(fig, ax) = plt.subplots()
fig
```

```python
# subplots
(fig, ax) = plt.subplots(nrows=1, ncols=2)

for eksen in ax:
    eksen.plot(np_arr1, np_arr2, "y")
    eksen.set_xlabel("x-ekseni")
    eksen.set_ylabel("y-ekseni")

plt.tight_layout()
plt.show()
```

```python
new_fig = plt.figure(figsize=(8,5))

new_ax = new_fig.add_axes([0.1,0.1,0.6,0.6])

new_ax.plot(np_arr1, np_arr1 ** 2, "r" ,label="x^2")
new_ax.plot(np_arr1, np_arr1 ** 3, "g" ,label="x^3")
new_ax.plot(np_arr1, np_arr1 ** (1/2), "b" ,label="x^1/2")
new_ax.set_xlabel("X")
new_ax.set_ylabel("Y")
new_ax.set_title("f(x)")

plt.legend() # loc=1 vs parametresi yeri değiştirir
plt.show()
```

```python
new_fig = plt.figure(dpi=100) # dpi kaliteyi belirtilir.

new_ax = new_fig.add_axes([0.1,0.1,0.6,0.6])

new_ax.plot(np_arr1, np_arr1 ** 2, "r" ,label="x^2")
new_ax.plot(np_arr1, np_arr1 ** 3, "g" ,label="x^3")
new_ax.set_xlabel("X")
new_ax.set_ylabel("Y")
new_ax.set_title("f(x)")

plt.legend() # loc=1 vs parametresi yeri değiştirir
plt.show()
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(np_arr1, np_arr2, color = "#3A95A8", linewidth=5.0)
benimEksen.plot(np_arr2, np_arr1, color = "#C96F23", alpha=0.5)
benimEksen.plot(np_arr2, np_arr2 + 2, color = "red", linestyle=":")
benimEksen.plot(np_arr2, np_arr2 + 20, color = "#000000", marker="o", markersize = 5, markerfacecolor="red")
plt.show()

# alpha -> saydamlık
# linewidth -> kalınlık
# linestyle -> stili değiştirir = "--" ":" "-."
# marker -> data point  = "o" "+" "x"
```
